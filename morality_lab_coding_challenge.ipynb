## **Python Challenge: Let's Debug!**

The following code packages all data transformation processes into one function. However, calling the function as it is currently written outputs only 11 of the 12 stories. How would you fix this error? Debug the following function, `q1()`, to ensure that its printed output includes all 12 stories.

````python
def q1():
    """
    Randomizes stimulus presentation order
    :returns: Nothing, iteratively prints stimuli in random order
    """
    causality_q = []
    prevent_q = []
    scenario_text = []

    # Read in the stimulus data using pandas
    stimuli_df = pd.read_csv(data_url)

    for index, row in stimuli_df.iterrows():
        causality_q.append(
            f"Did you cause {row['root']}?")
        prevent_q.append(
            f"Could you have prevented {row['gerund']}?")
        scenario_text.append(
            f"{row['Background']} {row['Intentional version']}")

    # Initialize a dictionary to hold the stimulus questions
    questions = {
        'causality': causality_q,
        'preventability': prevent_q,
        'scenario_text': scenario_text
    }

    # Make a permutation to display questions in random order
    q_order = np.random.permutation(range(1, len(causality_q), 1))

    # Print the questions according to that order
    for i in q_order:
        print(f'Story Number {i+1}:')
        print(questions['scenario_text'][i])
        print(questions['causality'][i])
        print(questions['preventability'][i])
        print('\n')

# Call the function
q1()
````
